```python
def inferieur(l, n):
    m = []
    alpha = 0
    for i in range(len(l)):
        for j in range(len(l[i])):
            if l[i][j] < n:
                alpha = l[i][j]
        m.append(alpha)
    return m
#2 is the index of the list in which 16 is in
#m = inferieur(list(indexes(bar(L)).values())[:2], 16)

#m
```

```python
def superieur(l, n):
    blist = [n]
    alpha = 0
    for i in range(len(l)):
        for j in range(len(l[i])):
            if l[i][j] == n+1:
                alpha = l[i][j]
                blist.append(alpha)
        n += 1
        
    return blist
#3 is the index of the next list in which 16 is in
#nb = superieur(list(indexes(bar(L)).values())[3:], 16)
#nb
```

```python
import re
def sub_function1(l):
    l1=[]
    for i in range(len(l)):
        l1.append(re.findall("\W", l[i]))
    return l1
```

```python
def sub_function2(l):
    l2=[]
    for i in range(len(l)):
        l1 = [l[i][j] for j in range(len(l[i])) if l[i][j]!=" " if l[i][j]!="."]
        l2.append(l1)
    for i in range(len(l2)):
        
        if l2[i].count('=')!=0:
            m = l2[i].index('=')
            l2[i][m-1] = l2[i][m-1] + l2[i][m]
            l2[i].remove('=')
        
    return l2
```

```python
def sub_function3(l):
    l2=[]
    for i in range(len(l)):
        l2.append("".join(l[i]))
    return l2
```
